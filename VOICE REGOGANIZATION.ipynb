```python
import os
import librosa
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

dataset_path = "C:\\Users\\haris\\Downloads\\Audio Data"
audio_files = [os.path.join(dataset_path, f) for f in os.listdir(dataset_path) if f.endswith('.wav')]

print(f"Found {len(audio_files)} audio files.")

if not audio_files:
    print("No .wav files found in the dataset directory.")
else:
    # Feature extraction: MFCC
    def extract_mfcc(file_path, n_mfcc=13):
        y, sr = librosa.load(file_path, sr=None, duration=5.0)
        mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
        return np.mean(mfcc, axis=1)

    # Pitch estimation (average fundamental frequency)
    def estimate_pitch(file_path):
        y, sr = librosa.load(file_path, sr=16000, duration=5.0)

     f0, voiced_flag, _ = librosa.pyin(y, fmin=50, fmax=400)
        # Average only the voiced parts
        return np.nanmean(f0)

    # Extract features
    features = np.array([extract_mfcc(f) for f in audio_files])
    pitches = np.array([estimate_pitch(f) for f in audio_files])
    print(f"Features shape: {features.shape}")
    print(f"Pitches shape: {pitches.shape}")

    if features.size > 0:
        features_scaled = StandardScaler().fit_transform(features)
        kmeans = KMeans(n_clusters=2, random_state=0, n_init=10)
        clusters = kmeans.fit_predict(features_scaled)

        # Determine which cluster is male/female based on average pitch
        cluster_pitch = {}
        for cluster_id in np.unique(clusters):
            cluster_pitch[cluster_id] = np.mean(pitches[clusters == cluster_id])

        # Male = lower pitch, Female = higher pitch
        male_cluster = min(cluster_pitch, key=cluster_pitch.get)
        female_cluster = max(cluster_pitch, key=cluster_pitch.get)

        print("\nResults:")
        for f, c in zip(audio_files, clusters):
            label = "Male" if c == male_cluster else "Female"
            print(f"File: {os.path.basename(f)} - Cluster: {c} - Predicted: {label}")
```
